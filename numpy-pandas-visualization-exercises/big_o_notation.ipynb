```python
import matplotlib.pyplot as plt
import math
```

```python
plt.figure(figsize = (8, 8))

x = list(range(1, 8))
y = [0*n + 1 for n in x]
plt.plot(x,y, c='tab:red', label='$O(1)$')

y2 = [math.log(n) for n in x]
plt.plot(x,y2, c='tab:blue', label='$O(log n)$')

y3 = x
plt.plot(x,y3, c='tab:green', label='$O(n)$')

y4 = [n*math.log(n) for n in x]
plt.plot(x,y4, c='tab:orange', label='$O(n log n)$')

y5 = [n**2 for n in x]
plt.plot(x,y5, c='tab:purple', label='$O(n^2)$')

y6 = [2**n for n in x]
plt.plot(x,y6, c='tab:brown', label='$O(2^n)$')

y7 = [math.factorial(n) for n in x]
plt.plot(x,y4, c='tab:pink', label='$O(n!)$')

y8 = [n**n for n in x]
plt.plot(x,y4, c='tab:cyan', label='$O(n^n)$')

plt.grid(True, ls = '--')
plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')

plt.legend()

plt.show
```
